```python
import argparse 

from pathlib import Path

parser = argparse.ArgumentParser()

parser.add_argument("-src", "--source",
                    type=str,
                    default='Hunter X Hunter',
                    help="File path of the source folder containing files",
                    metavar='')

args = parser.parse_args()


print(args.source)

print(args.source.strip(' " \\ . '))

source = args.source.strip(' " \\ . ')

print(source)

source = Path(source)

print(source)

print(source.exists())

print(source.resolve())

print(source.name)


def source_path(source):

    # Strip chracters aded by powershell
    # Maybe change to lstrip and r strip to avoid corrupting 
    source = args.source.strip(' " \\ . ')

    source = Path(source)
    
    # make sure the directory exists
    if source.exists() == False:
        print('Directory does not exist. Program terminated')
        exit()

    return source.name, source.resolve()

series_name, source_directory = source_path(args.source)

print(series_name, source_directory)
```
