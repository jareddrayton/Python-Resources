# Python 3 FizzBuzz Examples

```python
# FizzBuzz with a for loop

for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
# FizzBuzz with a list comprehension

lst = ["Fizz"*(not i%3) + "Buzz"*(not i%5) or str(i) for i in range(1, 101)]
for elem in lst:
    print(elem)
```
